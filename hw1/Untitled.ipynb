```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('lena.bmp')
plt.imshow(image)
plt.show()
```

```python
lena_1a = image[::-1]
plt.axis('off')
plt.imshow(lena_1a)
plt.savefig('1a.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
lena_1b = [row[::-1] for row in image]
plt.axis('off')
plt.imshow(lena_1b)
plt.savefig('1b.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
(h, w, d) = image.shape
lena_1c = np.array([ [image[j, i] for j in range(h)] for i in range(w) ])

# 顯示轉置後的圖像
plt.axis('off')
plt.imshow(lena_1c)
plt.savefig('1c.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
(h, w, d) = image.shape
center = (h/2,w/2)
angle = 45.0
scale = 1.0

# 旋轉
trans = cv2.getRotationMatrix2D(center, angle, scale)
lena_2a = cv2.warpAffine(image, trans, (w, h))
plt.axis('off')
plt.imshow(lena_2a)
plt.savefig('2d.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
(h, w, d) = image.shape
center = (h/2,w/2)
angle = 0.0
scale = 0.5

# 旋轉
trans = cv2.getRotationMatrix2D(center, angle, scale)
lena_2b = cv2.warpAffine(image, trans, (w, h))
plt.axis('off')
plt.imshow(lena_2b)
plt.savefig('2e.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
from PIL import Image

# Load the image using PIL
image = Image.open('lena.bmp')

# Convert the image to grayscale
gray_image = image.convert('L')

# Define the threshold
threshold = 128

# Binarize the image: Convert all pixels >= threshold to 255 (white), and < threshold to 0 (black)
binary_image = gray_image.point(lambda p: 255 if p >= threshold else 0)

# Save or display the binary image
binary_image.save('2f.bmp')
binary_image.show()
```
