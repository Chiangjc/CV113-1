```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image

image = cv2.imread('lena.bmp')
plt.imshow(image)
plt.axis('off')
plt.savefig('3a1.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
pixels = image.flatten()

hist = [0] * 256
for pixel in pixels:
    hist[pixel] += 1
    
print(hist)
```

```python
plt.bar(range(0, 256), hist)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('3a2.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
image = cv2.imread('lena.bmp')
image3 = image // 3 

plt.imshow(image3)  
plt.axis('off')
plt.savefig('3b1.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
pixels = image3.flatten() 
hist = [0] * 256
for pixel in pixels:
    hist[int(pixel)] += 1
    
print(hist)
```

```python
plt.bar(range(0, 256), hist)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('3b2.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
cdf = [0] * 256
cdf[0] = hist[0]
for i in range(1,256):
    cdf[i] = cdf[i - 1] + hist[i]

cdf_min = min(cdf)
cdf_max = max(cdf)
total_pixels = sum(hist)
# cdf_normalized = [(value - cdf_min) * 255 / (cdf_max - cdf_min) for value in cdf]
# total_pixels = sum(hist)
cdf_normalized = [i * 255 / total_pixels for i in cdf]
cdf_scaled = np.round(cdf_normalized).astype('uint8')
print(cdf_scaled)
```

```python
# 初始化均衡化後的圖像
equalized_image = np.zeros_like(image3)

# 逐像素映射
for i in range(image3.shape[0]):  # 遍歷行
    for j in range(image3.shape[1]):  # 遍歷列
        old_intensity = image3[i, j]  # 獲取原始像素值
        new_intensity = cdf_scaled[old_intensity]  # 映射到新強度值
        equalized_image[i, j] = new_intensity  # 更新均衡化後的圖像
```

```python
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')
plt.savefig('3c1.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
pixels = equalized_image.flatten()
pixels = [pixel for pixel in pixels]
hist = [0] * 256
for pixel in pixels:
    hist[int(pixel)] += 1
    
plt.bar(range(0, 256), hist)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('3c2.png', bbox_inches='tight', pad_inches=0)
plt.show()
```
