```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image as img

image = cv2.imread('lena.bmp')
plt.imshow(image)
plt.show()
```

```python
image = cv2.imread('lena.bmp', cv2.IMREAD_GRAYSCALE)
lena_arr = np.asarray(image)
```

```python
structure = [[-2, -1], [-2, 0], [-2, 1],
[-1, -2], [-1, -1], [-1, 0], [-1, 1], [-1, 2],
[0, -2],  [0, -1], [0, 0], [0, 1], [0, 2],
[1, -2],  [1, -1], [1, 0], [1, 1], [1, 2],
          [2, -1], [2, 0], [2, 1]]
```

```python
def dilation(lena_arr, structure):
    lena_dilation = np.zeros_like(lena_arr, dtype=np.uint8)
    for i in range(lena_arr.shape[0]):
        for j in range(lena_arr.shape[1]):
            
            max_value = 0
            for element in structure:
                new_i = i + element[0]
                new_j = j + element[1]
                if 0 <= new_i < lena_arr.shape[0] and 0 <= new_j < lena_arr.shape[1]:
                    max_value = max(max_value, lena_arr[new_i, new_j])
            lena_dilation[i, j] = max_value 
    return lena_dilation
```

```python
lena_dilation = dilation(lena_arr, structure)
img.fromarray(np.array(lena_dilation,dtype='uint8'))
```

```python
a = img.fromarray(np.array(lena_dilation,dtype='uint8'))
a.save("a.png")
```

```python
def erosion(lena_arr, structure):
    lena_erosion = np.zeros_like(lena_arr, dtype=np.uint8)
    for i in range(lena_arr.shape[0]):
        for j in range(lena_arr.shape[1]):
            
            min_value = 255
            for element in structure:
                new_i = i + element[0]
                new_j = j + element[1]
                if 0 <= new_i < lena_arr.shape[0] and 0 <= new_j < lena_arr.shape[1]:
                    min_value = min(min_value, lena_arr[new_i, new_j])
            lena_erosion[i, j] = min_value
    return lena_erosion
```

```python
lena_erosion=erosion(lena_arr, structure)
img.fromarray(np.array(lena_erosion,dtype='uint8'))
```

```python
b = img.fromarray(np.array(lena_erosion,dtype='uint8'))
b.save("b.png")
```

```python
lena_closing = erosion(dilation(lena_arr, structure), structure) 
img.fromarray(np.array(lena_closing,dtype='uint8'))
```

```python
c = img.fromarray(np.array(lena_closing,dtype='uint8'))
c.save("c.png")
```

```python
lena_opening = dilation(erosion(lena_arr, structure), structure)
img.fromarray(np.array(lena_opening,dtype='uint8'))
```

```python
d = img.fromarray(np.array(lena_opening,dtype='uint8'))
d.save("d.png")
```
